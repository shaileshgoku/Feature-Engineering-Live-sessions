```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
#combine df and df_country

final_df=pd.merge(df,df_country,on='Country Code', how = 'left')
```

# exploring dataset

```python
final_df.columns
```

```python
final_df.Country.value_counts()  #how many records are there in feature country
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
## pie chart
plt.pie(country_val,labels=country_names)
```

```python
## pie chart - Top 3 countries  that uses zomato
plt.pie(country_val[:3],labels=country_names[:3])
```

```python
## pie chart - Top 3 countries  that uses zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation:
Zomato maximum records or transaction are from india after that USA and then United Kingdoms


```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text'])
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

### Observations

1. Whenever Rating is between 4.5 to 4.9 --> Excellent

2. When Rating are between 4.0to 4.4 --> Very Good

3. When Rating are between 3.5 to 3.9 --> Good

4. When Rating are between 2.5 to 3.4 --> average

5. When Rating are between 1.8 to 2.8 --> poor


```python
ratings.head()
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color')
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette=['blue','red','yellow','green','green'])
```

### Observation:

1. Not Rated count is very high

2. Maximum no. of ratings are between 2.5 to 3.4
